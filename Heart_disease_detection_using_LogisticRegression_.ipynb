```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv('/content/heart_disease_data.csv')
```

```python
#1-->defective heart
#0-->healthy heart
X=df.drop(columns='target',axis=1)
Y=df['target']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
m=LogisticRegression()
```

```python
m.fit(X_train,Y_train)
```

```python
X_train_prediction=m.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction,Y_train)
```

```python
print("accuracy on training data: ",training_data_accuracy)
```

```python
X_test_prediction=m.predict(X_test)
testing_data_accuracy=accuracy_score(X_test_prediction,Y_test)
```

```python
print("Accuracy on testing Data:",testing_data_accuracy)
```

```python
input_data=(51,0,0,130,305,0,1,142,1,1.2,1,0,3)
input_data_numpy_array=np.asarray(input_data)
input_data_reshape=input_data_numpy_array.reshape(1,-1)
prediction=m.predict(input_data_reshape)
print(prediction)

if(prediction[0]==0):
  print("healthy heart")
else:
  print("defective heart")
```
